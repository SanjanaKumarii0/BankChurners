```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['Attrition_Flag'],axis=1)
y=df['Attrition_Flag']
```

```python
y.value_counts()
```

```python
y=y.map({'Existing Customer':0,'Attrited Customer':1})
display(y.head())
```

```python
y
```

```python
X=pd.get_dummies(X,columns=['Gender' ,'Education_Level','Dependent_count' , 'Marital_Status' ])                                                                                                                                                                                                                                                                                                                                                                                              
display(X.head())
```

```python
object_cols=X.select_dtypes(include='object').columns
X=pd.get_dummies(X,columns=object_cols)
display(X.head())
```

```python
X=X.astype(int)
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.naive_bayes  import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred ))
```

```python
import pickle 
with open ('model.pkl','wb')as fs:
   pickle.dump(model,fs)
```
